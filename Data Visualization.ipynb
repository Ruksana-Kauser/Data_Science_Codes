```python
from matplotlib import pyplot as plt
#import matplotlib.pyplot as plt
```

```python
month = ['Jan','Feb','Mar','Apr','May','Jun','Jul']
sales = [25000,22000,27000,20000,50000,67000,40000]
```

```python
month
```

```python
sales
```

```python
plt.figure(figsize=(15,6))
plt.plot(month,sales)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthwise Sales', size = 15)
```
